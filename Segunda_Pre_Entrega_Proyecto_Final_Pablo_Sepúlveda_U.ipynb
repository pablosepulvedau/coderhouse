## ELECCIÓN DE MODELO
Dada las características del dataset, se planea probar el siguiente modelo:
- Lineal Regression Multiple

Se planea evaluar el modelo de regresión lineal y establecer si es útil para predecir el costo de la vivienda. En los próximos entregables se demostrará si fue útil o no.

Las variables a utilizar para predecir RENT AMOUNT (R$), serán:
- City
- Area
- Rooms
- Bathroom
- Parking Spaces


```python
# Se importan las librerías NUMPY y SKLEARN
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

sns.set_style('darkgrid')
```

```python
# El gráfico resume como se comportan las variables numéricas en función de la ciudad.
g=sns.pairplot(df, hue='city',diag_kind='hist')
```

```python
for ax in g.axes.flat:
  plt.setp(ax.get_xticklabels(),rotation=45)
```

```python
# Se definen las variables usadas para predecir el valor de RENT AMOUNT (R$).
from sklearn.preprocessing import OneHotEncoder
area = df['area'].values
bathroom = df['bathroom'].values
rent_amount = df['rent amount (R$)'].values
rooms = df['rooms'].values
parking_spaces = df['parking spaces'].values
city = df['city'].values
```

```python
# La variable X contiene las variables independientes.
X = np.array([area, bathroom, rooms, parking_spaces,city]).T
```

```python
# La variable Y contiene la variable dependiente a predecir.
Y = np.array(rent_amount)
```

```python
# Se entrena el modelo de regresión lineal.
reg = LinearRegression()
reg = reg.fit(X,Y)
```

```python
# Se entrena el predictor con las variables de X
Y_pred = reg.predict(X)
```

```python
# Se establece el error del modelo usando MSE
error = np.sqrt(mean_squared_error(Y, Y_pred))
```

```python
# Se calcula el coeficiente de determinación.
r2 = reg.score(X,Y)
```

### TESTEO DEL MODELO
Para evaluar el modelo encontrado, se asignan valores a las variables AREA, BATHROOM, ROOMS, PARKING SPACES y CITY.

La ciudad se definió usando los siguientes datos:
- Sao Paulo = 1
- Rio de Janeiro = 2
- Campinas = 3
- Belo Horizonte = 4
- Porto Alegre = 5

```python

print('El error es: ',error)
print('El valor de r2 es: ', r2)
print("Los coeficientes son: \n'", reg.coef_)
area = 100
bathroom = 1
rooms = 2
parking_spaces = 1
city = 1
```

```python
print('El costo de la vivienda dada sus características es de: ', reg.predict([[area, bathroom, rooms, parking_spaces, city]]))
```

### CONCLUSIÓN TESTEO MODELO
El modelo posee un coeficiente de correlación de 0.49, por lo tanto, es de baja confiabilidad. Se sugiere probar con otro método tales como Decision Tree o Random Forest y evaluar si posee R2 mayor a 0.49.
